```python
import requests as re
from bs4 import BeautifulSoup
```

```python
site_url = "http://dtu.ac.in"
```

```python
site = re.get(site_url)
```

```python
site_main = BeautifulSoup(site.text, "html.parser")
```

```python
news = site_main.find_all("div", {"class": "latest_tab"})
```

```python
links = {}
cnt = 0

for i in news:
    links[cnt] = i.find_all("a")
    cnt+=1
```
